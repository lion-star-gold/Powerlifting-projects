```python
import matplotlib.pyplot as plt
```

```python
participation_numbers = pd.DataFrame(pd.to_datetime(df['Date']).dt.year.value_counts())
participation_numbers.sort_values(by = 'Date', inplace = True)
```

```python
plt.bar(x = participation_numbers.index,
          height = participation_numbers.Date,
       tick_label = participation_numbers.index,
        color = '#dc0714'
       )
plt.xticks(rotation = -90)
plt.xlabel("Year")
plt.ylabel("Number of registered British Powerlifting competitors")
plt.suptitle("British Powerlifting competitors over time")
#plt.show()
```

Shockingly low amount of competitors prior to 2008, with none in 2007. According to <a href = 'britishpowerlifting.org/board-directors'>this page</a>, the federation was founded in it's current form in 2008. Previous entries may be labelled incorrectly years after the event.

```python
df = df[df['Date'].dt.year > 2007]
```

How does women's participation vary over the years?

```python
df['Year'] = df['Date'].dt.year
```

```python
df_group = df.groupby(by = ['Sex', 'Year'])['Name'].count()
unique_years = list(df['Year'].unique())
unique_years.sort()
women_participation_count = df_group[0:14]
#women_participation_count
men_participation_count = df_group[14:28]
#men_participation_count
```

```python
plt.barh(y = unique_years, width = women_participation_count, color = '#002d59', edgecolor = 'black')
plt.barh(unique_years, men_participation_count, left = women_participation_count, color = '#dc0714', edgecolor = 'black')

plt.xticks(rotation = -90)
plt.legend(['Women', 'Men'], loc = 'lower right')
plt.suptitle("Number of competitors in all classic British Powerlifting competitons - national and local level")
plt.xlabel('# competitors')
plt.ylabel('Year')

#plt.show()
```

Above is confusing - create a table showing % participation by year.

```python
total_participants = df.groupby(by = 'Year')['Name'].count()
male_participants = df[df['Sex'] == 'M'].groupby(by = 'Year')['Name'].count()
female_participants = df[df['Sex'] == 'F'].groupby(by = 'Year')['Name'].count()
print("Are the three calculation lists the same length?", len(total_participants) == len(male_participants) == len(female_participants))

male_percent = round((male_participants / total_participants)*100, 1)
female_percent = round((female_participants / total_participants)*100, 1)
```

```python
participation = pd.DataFrame({'% Men participation': male_percent, '% Women participation': female_percent})
#participation.drop(participation.index, inplace=True) 
```

```python
participation
```

```python
plt.plot(participation.index.values, participation['% Men participation'], label = 'Men percent', marker = 'o', color = '#dc0714')
plt.plot(participation.index.values, participation['% Women participation'], label = 'Women percent', marker = 'o', color = '#002d59')
plt.xticks(ticks = participation.index.values, rotation = -45)
plt.suptitle("Percentages of men and women competing in British Powerlifting (classic) at all levels over time")
plt.xlabel("Year")
plt.ylabel("Percentage participation")
plt.legend()

plt.show()
```

2008 had a relatively large amount of female competitors - however we can see from the bar chart a few lines up that 2008 also has a relatively tiny amount of competitors total. So, excluding 2008, overall we see an increase in the relative amount of women competing, approaching around 40 % in the years 2017 - 2020 (funnily enough, these were also the years I competed).

The bar chart above shows a massive increase in the number of competitors over the 2013 - 2019 time period. Is this accompanied by an increase in the # of competitions?

How many national competitions (Federation == 'BP') were conducted per year, and was this impacted by COVID-19 disruption?
***

```python
BP_comps = df[df['Federation'] == 'BP']
list(BP_comps['MeetName'].unique())
```

```python
BP_comps[['MeetName','Year']].groupby('Year').count() #this gives the number of competitiors in national comps that year.
```

```python
national_comps = BP_comps.groupby(by = ['MeetName', 'Year'])['Year'].count()
national_comps = national_comps.to_frame()
national_comps = national_comps.unstack()
```

```python
#national_comps
```

The uppermost 'Year' isn't required and can be dropped using .droplevel().

```python
national_comps.columns = national_comps.columns.droplevel()
national_comps.reset_index(inplace = True)
national_comps
```

```python
national_comps.describe()
```

The above .describe() tells me that through the years, barring COVID-19 related disruption, the number of national-level 'classic' full-power competitions tends to increase, and so does the mean amount of competitors. Please note, the dataset includes competitors who compete 'classic' in 'equipped' competitions, which is competition legal but uncommon, and this can skew the results.

```python
drop_equipped_competitions = ['British Equipped Championships', 'British Equipped Powerlifting Championships', 'British Senior & Junior Equipped Powerlifting Championships']

national_comps = (national_comps[~national_comps['MeetName'].isin(drop_equipped_competitions)])
```
